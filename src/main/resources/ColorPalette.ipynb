```python
import matplotlib.pyplot as plt
```

```python
color_palette1 = {
    "Text": "#F5F5F5",           # Soft white, ensuring high contrast for readability
    "Background": "#2B2E3A",    # Deep gray-blue, evoking a dark, mystical atmosphere
    "Primary": "#5C946E",       # Mystical green for key elements like relics
    "Secondary": "#414242",     # Deep shadow gray for muted tiles and echoes
    "Accent": "#F4C542",        # Golden light to highlight relics or key UI
}

color_palette2 = {
    "Text": "#E0D8B0",           # Old scroll beige for medieval-style text
    "Background": "#2A2E38",    # Dark navy for a mystical and arcane setting
    "Primary": "#56A3D8",       # Ethereal blue for echoes and spectral elements
    "Secondary": "#7B7F86",     # Iron gray to add a medieval tone
    "Accent": "#D4A017",        # Golden relics to draw attention to treasures
}

color_palette3 = {
    "Text": "#E8E6E3",           # Soft parchment white for a relic-inspired text style
    "Background": "#1C1C1C",    # Charcoal black for a mysterious shadowy feel
    "Primary": "#644B7D",       # Arcane purple for mystical highlights
    "Secondary": "#9C9A97",     # Castle stone gray to ground the medieval tone
    "Accent": "#F4B400",        # Bright gold to highlight relics and key visuals
}
```

## Visual Swatch

```python
def show_color_palette(color_palette):
    fig, ax = plt.subplots(figsize=(20, 2))

    for i, (name, hex_code) in enumerate(color_palette.items()):
        ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=hex_code))
        ax.text(i + 0.5, -0.5, name, ha='center', va='center', fontsize=12)

    ax.set_xlim(0, len(color_palette))
    ax.set_ylim(0, 1)
    ax.axis('off')
    plt.title("Echo Relics Color Palette", fontsize=14, pad=10)
    plt.show()
```

```python
for color_palette in [color_palette1, color_palette2, color_palette3]:
    show_color_palette(color_palette)
```
